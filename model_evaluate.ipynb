```python
from ultralytics import YOLO
import yaml
import os
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
MODEL_PATH = "/content/drive/MyDrive/boneimages/yolo_runs/run1/weights/best.pt"
DATA_YAML = "/content/drive/MyDrive/boneimages/data_fixed.yaml"
```

```python
model = YOLO(MODEL_PATH)

with open(DATA_YAML) as f:
    data = yaml.safe_load(f)

CLASS_NAMES = data['names']
```

```python
print("Evaluating model on validation dataset...")
results = model.val(
    data=DATA_YAML,
    imgsz=640,
    batch=16
)

metrics = results.results_dict

```

```python
print("\n=== Overall Metrics ===")
print(f"Precision: {metrics['metrics/precision(B)']:.4f}")
print(f"Recall:    {metrics['metrics/recall(B)']:.4f}")
print(f"mAP50:     {metrics['metrics/mAP50(B)']:.4f}")

#f1 score
precision = metrics['metrics/precision(B)']
recall = metrics['metrics/recall(B)']

f1 = 2 * (precision * recall) / (precision + recall + 1e-6)
print(f"F1-score:  {f1:.4f}")
```
